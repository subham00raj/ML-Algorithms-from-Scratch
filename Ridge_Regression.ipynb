```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_ = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/USA_Housing.csv')
df_.drop('Address',inplace = True, axis=1)
df_.head()
```

```python
from sklearn.preprocessing import StandardScaler

ss = StandardScaler()
df = pd.DataFrame(ss.fit_transform(df_.drop('Price',axis=1)), columns = df_.columns[:5])
df['Price'] = df_['Price']
df.head()
```

```python
X = df[['Avg. Area Income', 'Avg. Area House Age', 'Avg. Area Number of Rooms','Avg. Area Number of Bedrooms', 'Area Population']]
y = df['Price']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_squared_error, r2_score

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.3, random_state = 42)

model = Ridge(alpha = 0.1)
model.fit(X_train,y_train)
y_pred = model.predict(X_test)

print('Coefficients :',model.coef_, '\n' 'Intercept :',model.intercept_)
print('MSE :',mean_squared_error(y_test,y_pred))
print('R2 Score :',r2_score(y_test,y_pred))
```

## Inverse Method

```python
class Ridge:

  def __init__(self,alpha):
    self.y_pred = None
    self.coef_ = None
    self.intercept_ = None
    self.alpha = alpha

  def fit(self,X_train,y_train):

    X_train = np.hstack((np.ones((X_train.shape[0],1)) , X_train))
    I = np.identity(X_train.shape[1])
    I[0][0] = 0 # bias term is not regularised
    W = np.linalg.inv(X_train.T @ X_train + self.alpha * I) @ (X_train.T @ y_train)

    self.coef_ = W[1:]
    self.intercept_ = W[0]

  def predict(self,X_test):
    self.y_pred = np.dot(self.coef_, X_test.T) + self.intercept_
    return self.y_pred

  @staticmethod
  def mean_squared_error(y_test,y_pred):
    return ((y_test - y_pred)**2).mean()

  @staticmethod
  def r2_score(y_test,y_pred):
    y_mean = np.mean(y_test)
    ss_total = np.sum((y_test - y_mean) ** 2)
    ss_residual = np.sum((y_test - y_pred) ** 2)
    r2 = 1 - (ss_residual / ss_total)
    return r2


model = Ridge(0.1)
model.fit(X_train,y_train)
y_pred = model.predict(X_test)


print('Coefficients :',model.coef_, '\n' 'Intercept :',model.intercept_)
print('MSE :',model.mean_squared_error(y_test,y_pred))
print('R2 Score :',model.r2_score(y_test,y_pred))
```

## Gradient Descent

```python
class GD_Ridge_Regressor:

  def __init__(self, learning_rate, epochs, alpha):
    self.learning_rate = learning_rate
    self.epochs = epochs
    self.coef_ = None
    self.intercept_ = None
    self.alpha = alpha
    self.y_pred = None
    self.cost_history = []


  def fit(self,X_train,y_train):

    self.coef_ = np.zeros((X_train.shape[1],))
    self.intercept_ = 0

    for epoch in range(self.epochs):

      y_hat = np.dot(self.coef_ , X_train.T) + self.intercept_

      cost = np.sum((y_train - y_hat)**2) + self.alpha * (np.dot(self.coef_.T,self.coef_)) / X_train.shape[0]
      self.cost_history.append(cost)

      dj_dw = (np.dot((y_hat - y_train),X_train) + self.alpha * self.coef_) / X_train.shape[0]
      dj_db = -2 * np.mean(y_train - y_hat)

      self.coef_ = self.coef_ - self.learning_rate * dj_dw
      self.intercept_ = self.intercept_ - self.learning_rate * dj_db

    return self.coef_ , self.intercept_

  def plot(self):
    x = np.arange(0,len(self.cost_history),1)
    plt.plot(x,self.cost_history)
    plt.xlabel('Epochs')
    plt.ylabel('Cost')

  def predict(self,X_test):
    self.y_pred = np.dot(self.coef_, X_test.T) + self.intercept_
    return self.y_pred

  @staticmethod
  def mean_squared_error(y_test,y_pred):
    return ((y_test - y_pred)**2).mean()

  @staticmethod
  def r2_score(y_test,y_pred):
    y_mean = np.mean(y_test)
    ss_total = np.sum((y_test - y_mean) ** 2)
    ss_residual = np.sum((y_test - y_pred) ** 2)
    r2 = 1 - (ss_residual / ss_total)
    return r2


model = GD_Ridge_Regressor(learning_rate = 0.1, epochs = 500, alpha = 0.1)
model.fit(X_train,y_train)
y_pred = model.predict(X_test)


print('Coefficients :',model.coef_, '\n' 'Intercept :',model.intercept_)
print('MSE :',model.mean_squared_error(y_test,y_pred))
print('R2 Score :',model.r2_score(y_test,y_pred))
```
